```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# Parameters
n_classes = 3
colors = ["b","g","y"]

# Load data
f = open("dectree.csv")
# f.readline()  # if there is the header
data = np.loadtxt(f, delimiter=',')
print(data)
```

```python
X = data[:,:2] #récupère tous les lignes des colonnes 0 à 1
y = data[:,2]

print(X)
print(y)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
clf = DecisionTreeClassifier().fit(X, y)  #entraine le modèle

x_min, x_max = 0,100
y_min, y_max = 0,100
plot_step = 10
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])  #ravel renvoie un tableau aplati contigu
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.viridis)  #contourf trace des contours remplis

for i in range(0,3):   
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=colors[i], cmap=plt.cm.viridis)

plt.show()
```
