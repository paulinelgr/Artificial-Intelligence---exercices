```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f= open("regression.csv")
data = np.loadtxt(f,delimiter=",")
X = data[:,0]
Y = data[:,1]
```

```python
print(data)
```

```python
plt.scatter(X,Y, color='black')
```

# Create the Neural Network

```python
from keras.models import Sequential  #for multi layer network
from keras.layers.core import Dense  # for dense layers (fully connected)
```

## 1. Prepare the model

```python
model = Sequential()
model.add(Dense(units = 1, activation = 'linear', input_dim=1))
```

# 2. Compile the model (create the necessary memory)

```python
model.compile(loss='mse', optimizer='adam')   #loss c'est la fonction objectif/l'écart entre la data et le modèle  (ici mse: somme des moindres carrés)
model.summary()
```

# 3. Train the model

```python
model.fit(X, Y, epochs=500, verbose=1)   #loss diminue au fil des epochs
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='black')
plt.scatter(X, Y_pred, color='blue')
```

```python
for layer in model.layers:
    weights = layer.get_weights()
    print(weights)
```
